```python
# ============================
# STEP 1: Install Required Libraries
# ============================
!pip install transformers accelerate sentencepiece --quiet

# ============================
# STEP 2: Import Libraries
# ============================
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# ============================
# STEP 3: Load Lightweight Flan-T5 Model
# ============================
model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Create pipeline
chatbot = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

# ============================
# STEP 4: Define Health Bot Logic
# ============================
def ask_health_bot(user_input):
    prompt = f"""
You are a helpful and friendly health assistant.
Answer general health-related questions clearly, but do not give medical diagnoses or prescriptions.
If a question is too personal or dangerous, say:
"I'm here to help with general information, but please consult a licensed doctor for personal advice."

Question: {user_input}
Answer:"""

    response = chatbot(prompt, max_new_tokens=150)[0]['generated_text']
    return response.strip()

# ============================
# STEP 5: Chat Loop
# ============================
print("🤖 Flan-T5 HealthBot is ready! Ask your health questions.")
print("Type 'exit' to quit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("HealthBot: Take care! 👋")
        break
    reply = ask_health_bot(user_input)
    print("HealthBot:", reply, "\n")
```
