```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.show()
```

```python
sns.pairplot(df,hue='Species')
plt.suptitle('Pairplot',y=0.01)
plt.show()
```

```python
df.hist(figsize=(10,8),edgecolor='black')
plt.suptitle('Histogram of Numerical Features',fontsize=18)
plt.tight_layout()
plt.show()  
```

```python
plt.figure(figsize=(10,10))
sns.histplot(data=df,x='SepalLengthCm',bins=20,color='red')
plt.title('Sepal Length Distribution')
plt.xlabel('sapel')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['SepalLengthCm'],color='skyblue')
plt.title('boxplot')
plt.xlabel('sepallenth')
plt.ylabel('cm')
plt.show()
```

```python
df_numeric=df.select_dtypes(include='number')
plt.figure(figsize=(10,6))
sns.boxplot(data=df_numeric)
plt.title('Boxplot of Numeric Columns')
plt.show()
```
